```python
import numpy as np
import pandas as pd
from pandas import read_csv
dfLaliga = pd.read_csv("Laliga.csv")
dfLaliga = dfLaliga.replace('-',0)
dfLaliga
```

```python
#Print all the teams which have started playing between 1930-1980. (5 points)
```

```python
df = dfLaliga[(dfLaliga["Debut"]>="1930") & (dfLaliga["Debut"]<="1980")]
print(df)
```

```python
#Print the list of teams which came Top 5 in terms of points (3 points)
dfLaliga.Points = pd.to_numeric(dfLaliga['Points'])
dfLaliga.nlargest(5,"Points")
```

```python
#Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference. (7 points)
dfLaliga.GoalsFor = pd.to_numeric(dfLaliga['GoalsFor'])
dfLaliga.GoalsAgainst = pd.to_numeric(dfLaliga['GoalsAgainst'])
def Goal_diff_count():
    return[dfLaliga["GoalsFor"]-dfLaliga["GoalsAgainst"]]
print(Goal_diff_count())
print('max goal diff:',np.max(df))
print('min goal diff:',np.min(df))
```

```python
# Create a new column with name â€œWinning Percentâ€ and append it to the data set (6 points)
# If there are any numerical error, replace it with 0%
dfLaliga.GamesWon = pd.to_numeric(dfLaliga['GamesWon'])
dfLaliga.GamesPlayed = pd.to_numeric(dfLaliga['GamesPlayed'])
dfLaliga['Winning Percent'] =  (dfLaliga['GamesWon'] / dfLaliga['GamesPlayed'])*100
dfLaliga = dfLaliga.fillna({"Winning Percent":0})
dfLaliga
```

```python
# Print the top 5 teams which has the highest Winning percentage
dfLaliga.nlargest(5,"Winning Percent")['Team']
```

```python
# Group teams based on their â€œBest positionâ€ and print the sum of their points for all positions (6 points)
dfLaliga.BestPosition = pd.to_numeric(dfLaliga['BestPosition'])
dfLaliga.Points = pd.to_numeric(dfLaliga['Points'])
dfLaliga.groupby("BestPosition")["Points"].agg("sum")
```
